```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('emails.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, accuracy_score, precision_recall_curve, RocCurveDisplay, PrecisionRecallDisplay, auc, roc_auc_score



def report(classifier):
    classifier.fit(xtrain, ytrain)
    ypred = classifier.predict(xtest)

    cm = confusion_matrix(ypred, ytest)
    display = ConfusionMatrixDisplay(cm)
    display.plot()

    
    precision, recall, threshold = precision_recall_curve(ytest, ypred)
    display = PrecisionRecallDisplay(precision=precision, recall=recall)
    display.plot()

    fpr, tpr, threshold = roc_curve(ytest, ypred)
    rocauc = auc(fpr, tpr)
    display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=rocauc)
    display.plot()

    
```

```python

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=15)
report(knn)
```

```python
from sklearn.svm import SVC
svc = SVC(gamma='auto')
report(svc)
```
