```python
df = df.drop(columns=['CUSTOMERNAME', 'PHONE', 'ADDRESSLINE1', 'ADDRESSLINE2', 'CONTACTFIRSTNAME', 'CONTACTLASTNAME', 'QTR_ID', 'CITY', 'STATE', 'TERRITORY', 'PHONE'])
```

```python
df = df.drop(columns=['ORDERDATE', 'STATUS', 'YEAR_ID', 'MONTH_ID', 'ORDERNUMBER'])
```

```python
df = df.drop(columns=['POSTALCODE'])
```

```python
df.head()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

def convert_cats(col):
    df[col] = encoder.fit_transform(df[col].values)
```

```python
cols = ['PRODUCTLINE', 'PRODUCTCODE', 'COUNTRY', 'DEALSIZE']
```

```python
for col in cols:
    convert_cats(col)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df = scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 16):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
print(wcss)
```

```python
k = list(range(1,16))

plt.figure()
plt.xlabel('k value')
plt.ylabel('inertia')
# plt.grid()
plt.plot(k, wcss)
plt.show()
```

```python
# sampledf = df[['QUANTITYORDERED', 'PRICEEACH']]

sdf = pd.read_csv('sales_data_sample.csv', encoding='unicode escape')
sdf = sdf[['QUANTITYORDERED', 'PRICEEACH']]
sdf.isna().sum()

sdf = scaler.fit_transform(sdf)
sdf
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, random_state=10, n_init=10)
clusters = kmeans.fit_predict(sdf)
```

```python
plt.scatter(sdf[:,0], sdf[:,1], c=clusters, cmap='viridis')
```
