```python
# function : y = (x+3) ^ 2
# Start from x = 2
```

```python
current_x = 2
rate = 0.01
delta_x = 1
precision = 0.000001
iteration_counter = 0
max_iteration = 10000
```

```python
def slope(x):
    return ( 2 * (x + 3) )

def value_y(x):
    return ( (x + 3) ** 2 )


x = []
y = []
x.append(current_x)
y.append(value_y(current_x))
```

```python
while delta_x > precision and iteration_counter < max_iteration:
    prev_x = current_x
    current_x = prev_x - rate * slope(prev_x)
    print(f"X is : {current_x}")

    x.append(current_x)
    y.append(value_y(current_x))

    delta_x = abs(prev_x - current_x)

    print(f"Iteration Count : {iteration_counter}")
    print(f"X is : {current_x}")

print(f"Found the local minima at : {current_x}")
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('y = (x + 3) ^ 2')
plt.scatter(x, y)
plt.show()
```
