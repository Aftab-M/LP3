```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
plt.xlabel('Exited')
plt.ylabel('Count')
df['Exited'].value_counts().plot.bar()
plt.show()
```

```python
df['Geography'].value_counts()
```

```python
df = pd.concat([df, pd.get_dummies(df['Geography'], prefix='Geo').astype(int)], axis=1)
df.head()
```

```python
df = pd.concat([df, pd.get_dummies(df['Gender']).astype(int)], axis=1)
df.head()
```

```python
df.drop(columns=['RowNumber', 'CustomerId', 'Surname', 'Geography', 'Gender'], inplace=True)
df.head()
```

```python
x = df.loc[:, df.columns != 'Exited'].values
y = df['Exited'].values
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, random_state=20)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
xtrain.shape
```

```python

import tensorflow as tf
from tensorflow.keras.layers import Dense, Conv1D, Flatten, Input
from tensorflow.keras.models import Sequential, Model
```

```python
model = Sequential()
model.add(Flatten(input_shape=(13,)))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
df.dtypes
```

```python
model.compile(optimizer='adam', metrics=['accuracy'], loss='BinaryCrossentropy')
```

```python
model.fit(xtrain, ytrain, batch_size=64, validation_split=0.1, epochs=100)
# print(xtrain.shape, ytrain.shape)
```

```python
pred = model.predict(xtest)
```

```python
ypred = []
for val in pred:
    if val > 0.5 :
        ypred.append(1)
    else :
        ypred.append(0)

```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
accuracy_score(ytest, ypred)
```

```python
cm = confusion_matrix(ytest, ypred)
display = ConfusionMatrixDisplay(cm)
display.plot()
```

```python
from sklearn.neural_network import MLPClassifier

nnclassifier = MLPClassifier(hidden_layer_sizes=(100), activation='logistic', max_iter=300)
nnclassifier.fit(xtrain, ytrain)
```

```python
ypred2 = nnclassifier.predict(xtest)
```

```python
accuracy_score(y_pred=ypred2, y_true=ytest)
```

```python
nnclassifier.score(xtest, ytest)
```
