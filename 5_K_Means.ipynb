```python
from sklearn.preprocessing import LabelEncoder
def convert_cats(col):
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].values)
```

```python
cols = ['PRODUCTLINE', 'PRODUCTCODE', 'COUNTRY', 'DEALSIZE']

for col in cols:
    convert_cats(col)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaled_df = scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans

wcss = []

for k in range(1,15):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state = 15)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)

```

```python
k = list(range(1,15))

plt.plot(k, wcss)

plt.xlabel('Clusters')
plt.ylabel('Scores')
plt.title('Finding the correct number of clusters w/ Elbow point method')
plt.grid()
plt.show()
```

```python
# AT K=4, THE GRAPH STARTS TO MOVE ALMOST PARALLEL TO X-AXIS
# THIS MEANS THAT IT IS THE OPTIMAL NUMBER OF CLUSTERS FOR THIS DATASET
```
